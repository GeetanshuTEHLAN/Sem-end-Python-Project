```python
# Question 1, Part(A)

def check_password(password):
    
    if len(password) < 9:
        return "Invalid: Password must be at least 9 characters long."

   
    if not any(char.isupper() for char in password):
        return "Invalid: Password must contain at least 1 uppercase letter."

    
    return "Valid: Password is secure."


user_password = input("Enter your password: ")


result = check_password(user_password)
print(result)
```

```python
# Question 1, Part(B)

def advanced_check_password(password):
    
    if len(password) < 9:
        return "Invalid: Password must be at least 9 characters long."

    
    if not any(char.isupper() for char in password):
        return "Invalid: Password must contain at least 1 uppercase letter."

    
    if not any(char.islower() for char in password):
        return "Invalid: Password must contain at least 1 lowercase letter."

    
    return "Valid: Password is secure."


user_password = input("Enter your password: ")


result = advanced_check_password(user_password)
print(result)
```

```python
# Answer (A)

import pandas as pd

# Loading the dataset 
wine_df = pd.read_excel(r"C:\Users\HPE\Downloads\Wine.xlsx")

# Displaying the last 50 records
last_50_records = wine_df.tail(50)
print("Last 50 Records:")
print(last_50_records)
```

```python
# Answer (B)

# List of top 5 Bordeaux wineries
top_wineries = ['Château Margaux', 'Château Latour', 'Château Lafite Rothschild', 'Château Cheval Blanc', 'Château Mouton Rothschild']

# Creating a dictionary to store dataframes for each winery
winery_dataframes = {}

# Filtering the data for each winery and storing it in the dictionary
for winery in top_wineries:
    winery_dataframes[winery] = wine_df[wine_df['winery'] == winery]

# Displaying the first few rows of the data for Château Margaux as an example
print("Data for Château Margaux:")
print(winery_dataframes['Château Margaux'].head())

# Now, let's export each winery's data to a separate Excel file
for winery, data in winery_dataframes.items():
    # Creating the filename
    filename = f"Q2B_{winery.replace(' ', '_')}.xlsx"
    
    # Exporting the data to an Excel file
    data.to_excel(filename, index=False)
    
    # Printing a message indicating that the data has been exported
    print(f"Data for {winery} exported to {filename}")
```

```python
# Answer (C)

# List of top 5 Bordeaux wineries
top_wineries = ['Château Margaux', 'Château Latour', 'Château Lafite Rothschild', 'Château Cheval Blanc', 'Château Mouton Rothschild']

# Creating a dictionary to store descriptive statistics for each winery
winery_statistics = {}

# Computing descriptive statistics for the 'price' column of each winery
for winery in top_wineries:
    winery_data = wine_df[wine_df['winery'] == winery]
    statistics = winery_data['price'].describe()
    winery_statistics[winery] = statistics

# Displaying the descriptive statistics for Château Margaux as an example
print("Descriptive Statistics for Château Margaux:")
print(winery_statistics['Château Margaux'])

# Now, let's export each winery's descriptive statistics to a separate Excel file
for winery, statistics in winery_statistics.items():
    # Creating the filename
    filename = f"Q2C_{winery.replace(' ', '_')}.xlsx"
    
    # Exporting the descriptive statistics to an Excel file
    statistics.to_excel(filename)
    
    # Printing a message indicating that the data has been exported
    print(f"Descriptive statistics for {winery} exported to {filename}")
```

```python
# Answer (D)

# Selecting wines from Burgundy
burgundy_data = wine_df[wine_df['province'] == 'Burgundy']

# Computing descriptive statistics for the 'price' and 'points' columns
burgundy_statistics = burgundy_data[['price', 'points']].describe()

# Displaying the descriptive statistics for Burgundy as an example
print("Descriptive Statistics for Burgundy:")
print(burgundy_statistics)

# Exporting the descriptive statistics to an Excel file
burgundy_statistics.to_excel("Q2D_Burgundy.xlsx")

# Printing a message indicating that the data has been exported
print("Descriptive statistics for Burgundy exported to Q2D_Burgundy.xlsx")
```

```python
# Answer (E)

# Selecting new-world wines (countries outside Europe)
new_world_data = wine_df[wine_df['land'] != 'Europe']

# Calculating the average price and points for each new-world country
new_world_statistics = new_world_data.groupby('country')[['price', 'points']].mean()

# Displaying the average statistics for new-world countries as an example
print("Average Statistics for New-World Countries:")
print(new_world_statistics)

# Exporting the average statistics to separate Excel files for each country
for country, statistics in new_world_statistics.iterrows():
    # Creating the filename
    filename = f"Q2E_{country.replace(' ', '_')}.xlsx"
    
    # Exporting the statistics to an Excel file
    statistics.to_excel(filename, index_label='Metric')
    
    # Printing a message indicating that the data has been exported
    print(f"Average statistics for {country} exported to {filename}")
```
